```python
## data preprocessing is complete and now various functions are prepared for processing

def entropy(X_data):
    
    y = np.unique(X_data, return_counts = True)
    N = X_data.shape[0]
    ent = 0.0
    for ix in y[1]:
        
        ent = ent + (-1.0 * (ix / N)*(np.log2(ix/N)))
       
    return ent
    
```

```python
def divide_data(X_data, fkey, fval):
    l = pd.DataFrame([], columns = X_data.columns)
    r = pd.DataFrame([], columns = X_data.columns)
    
    for ix in range(X_data.shape[0]):
        if X_data[fkey].loc[ix] <= fval:
            l = l.append(X_data.loc[ix])
        else:
            r = r.append(X_data.loc[ix])
    return l, r
```

```python
def information_gain(X_data, fkey, fval):
    left, right = divide_data(X_data, fkey, fval)
    
    l = float(left.shape[0])
    r = float(right.shape[0])
    N = float(X_data.shape[0])
    
    if l == 0 or r == 0:
        return -10000000
    
    ans = entropy(X_data.Survived) - ((l / N) * entropy(left.Survived)) - ((r / N) * entropy(right.Survived))
    return ans
```

```python
class DecisionTree:
    
    def __init__(self, depth = 0, max_depth = 5):
        
        self.left = None
        self.right = None
        self.max_depth = max_depth
        self.depth = depth
        self.target = None
        self.fkey = None
        self.fval = None
        
    def train(self,X_data):
        
        features = ["Pclass", "Sex","Age","SibSp","Parch","Fare"]
        i_gain = []
        for fx in features:
            i_gain.append(information_gain(X_data, fx, X_data[fx].mean()))
        
        self.fkey = features[np.argmax(i_gain)]
        self.fval = X_data[self.fkey].mean()
        
        print("making tree on the basis of {}".format(self.fkey))
        
        left,right = divide_data(X_data, self.fkey, self.fval)
        left = left.reset_index(drop = True)
        right = right.reset_index(drop = True)
        
        if left.shape[0] == 0 or right.shape[0] == 0:
            if X_data.Survived.mean() > 0.5:
                self.target = 1
            else:
                self.target = 0
            return
        
        if self.depth >= self.max_depth:
            if X_data.Survived.mean() > 0.5:
                self.target = 1
            else:
                self.target = 0
            return
        
        self.left = DecisionTree(depth = self.depth + 1, max_depth = self.max_depth)
        self.left.train(left)
        self.right = DecisionTree(depth = self.depth + 1, max_depth = self.max_depth)
        self.right.train(right)
        
        if X_data.Survived.mean() > 0.5:
            self.target = 1
        else:
            self.target = 0
        return
    
    def predict(self, test):
        
        if test[self.fkey] > self.fval:
            
            if self.right is not None:
                return self.right.predict(test)
            return self.target
        else:
            if self.left is not None:
                return self.left.predict(test)
            return self.target
        
    def accurracy(self, test):
        output = test.Survived
        correct_answer = 0
        for ix in range(test.shape[0]):
            pred = self.predict(test.loc[ix])
            if pred == output.loc[ix]:
                correct_answer = correct_answer + 1
        return float(correct_answer/ test.shape[0])
```
